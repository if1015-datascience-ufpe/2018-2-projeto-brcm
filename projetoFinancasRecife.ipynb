```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
end_despesas = './dados_prefeitura/recife-dados-despesas-2017.csv'
end_receitas = './dados_prefeitura/recife-dados-receitas-2017.csv'

# Dataframe com os dados das despesas
despesas = pd.read_csv(end_despesas, sep='\t',delimiter=';',decimal=',')
desp_df = pd.DataFrame(despesas)
desp_df.head()
```

```python
desp_df.dtypes
```

```python
# Excluir colunas redundantes
desp_colunas_excluir = ['orgao_codigo','unidade_codigo','categoria_economica_codigo','grupo_despesa_codigo','modalidade_aplicacao_codigo','elemento_codigo','subelemento_codigo','funcao_codigo','subfuncao_codigo','programa_codigo','acao_codigo','fonte_recurso_codigo','empenho_modalidade_codigo','credor_codigo','modalidade_licitacao_codigo']
despesas_df = desp_df.drop(desp_colunas_excluir, axis=1)

# Ajuste de tipos numericos
despesas_df['valor_liquidado'] = despesas_df.valor_liquidado.astype('float64')
despesas_df['valor_empenhado'] = despesas_df.valor_empenhado.astype('float64')
despesas_df['valor_pago'] = despesas_df.valor_pago.astype('float64')

print(despesas_df.dtypes)

despesas_df.isnull().sum()
```

```python
despesas_df.describe()
```

```python
# Dados da Receita da Prefeitura
rec = pd.read_csv(end_receitas, sep='\t',delimiter=';',decimal=',')
rec_df = pd.DataFrame(rec)
rec_df.head()
```

```python
print(rec_df.dtypes)
```

```python
# Excluir colunas redundantes
rec_colunas_excluir = ['mes','orgao_codigo','unidade_codigo','fonte_recurso_codigo','categoria_receita_codigo','receita_local_nome','fonte_origem_receita_codigo','subfonte_receita_codigo','rubrica_receita_codigo','alinea_receita_codigo','subalinea_receita_codigo','subalinea_receita_codigo','receita_local_codigo']
receita_df = rec_df.drop(rec_colunas_excluir, axis=1)

print(receita_df.dtypes)

receita_df.isnull().sum()
```

```python
receita_df.describe()
```

```python
# Gastos de acordo com sua função
gasto_por_orgao = despesas_df.groupby(['funcao_nome']).agg({'valor_pago':'sum'})

gasto_por_orgao
```

```python
# Substituir codigo por nomes de mes
meses = {1:'Janeiro',2:'Fevereiro',3:'Março',4:'Abril',5:'Maio',6:'Junho',7:'Julho',8:'Agosto',9:'Setembro',10:'Outubro',11:'Novembro',12:'Dezembro'}

mes_nome = []

for mes in despesas_df['mes_movimentacao']:
    mes_nome.append(mes)
    
lista = [meses.get(item,item) for item in mes_nome]

despesas_df['mes_movimentacao'] = lista

lista_meses = ['Janeiro','Fevereiro','Março','Abril','Maio','Junho','Julho','Agosto','Setembro','Outubro','Novembro','Dezembro']

# Analisar qual o mês com mais gasto

gasto_por_mes = despesas_df.groupby(['mes_movimentacao']).agg({'valor_pago':'sum'})

receita_por_mes = receita_df.groupby(['mes_nome']).agg({'receita_arrecadada':'sum'})

comp_df = pd.concat([gasto_por_mes,receita_por_mes],axis=1,sort=False)

grafico_com = comp_df.plot.bar()

grafico_com.set_title('Comparação mensal entre Receita e Despesa')
```

```python
# Verificar o saldo anual
gasto_total = gasto_por_mes['valor_pago'].sum()
receita_total = receita_por_mes['receita_arrecadada'].sum()
balanco = [gasto_total,receita_total]
balanco_total = pd.DataFrame(balanco)
graf_balanco = balanco_total.plot.bar()
graf_balanco.set_xticklabels(['Gastos','Receita'])
graf_balanco
```

```python
gasto_por_grupos = despesas_df.groupby(['grupo_despesa_nome']).agg({'valor_pago':'sum'})

gasto_por_grupos.plot.barh()
```
