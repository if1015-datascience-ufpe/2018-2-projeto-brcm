```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import re
import csv
```

```python
end_despesas = './dados_prefeitura/recife-dados-despesas-2017.csv'
end_receitas = './dados_prefeitura/recife-dados-receitas-2017.csv'

despesas = pd.read_csv(end_despesas, sep='\t',delimiter=';')
despesas_df = pd.DataFrame(despesas)
despesas_df.head()
```

```python
receita = pd.read_csv(end_receitas, sep='\t',delimiter=';')
receita_df = pd.DataFrame(receita)
receita_df.head()
```

```python
despesas_df['valor_empenhado'] = despesas_df['valor_empenhado'].str.replace(',','.')
despesas_df['valor_liquidado'] = despesas_df['valor_liquidado'].str.replace(',','.')
despesas_df['valor_pago'] = despesas_df['valor_pago'].str.replace(',','.')

despesas_df['valor_liquidado'] = despesas_df.valor_liquidado.astype('float64')
despesas_df['valor_empenhado'] = despesas_df.valor_empenhado.astype('float64')
despesas_df['valor_pago'] = despesas_df.valor_pago.astype('float64')

print(despesas_df.dtypes)

despesas_df.isnull().sum()
```

```python
gasto_por_orgao = despesas_df.groupby(['orgao_nome']).agg({'valor_pago':'sum'})

gasto_por_orgao

# gasto_por_orgao.plot.barh()
```

```python
gasto_por_mes = despesas_df.groupby(['mes_movimentacao']).agg({'valor_pago':'sum'})

gasto_por_mes

gasto_por_mes.plot.barh()
```
