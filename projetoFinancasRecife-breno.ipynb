# Analisar balanço mensal entre Receita e Despesa


```python
gasto_por_mes = despesas_df.groupby(['mes_movimentacao']).agg({'valor_pago':'sum'})

receita_por_mes = receita_df.groupby(['mes']).agg({'receita_arrecadada':'sum'})

comp_df = pd.concat([gasto_por_mes,receita_por_mes],axis=1,sort=False)

grafico_com = comp_df.plot.bar()

lista_meses = ['Janeiro','Fevereiro','Março','Abril','Maio','Junho','Julho','Agosto','Setembro','Outubro','Novembro','Dezembro']
grafico_com.set_title('Comparação mensal entre Receita e Despesa')
grafico_com.set_xticklabels(lista_meses)
```

## Visualizar o total das despesas e total da receita

```python
gasto_total = gasto_por_mes['valor_pago'].sum()
receita_total = receita_por_mes['receita_arrecadada'].sum()
balanco = [gasto_total,receita_total]
balanco_total = pd.DataFrame(balanco,\
                            columns=['valores'])
balanco_total

graf_balanco = balanco_total.plot(kind='bar', stacked=True, colormap='Spectral')

graf_balanco.set_xticklabels(['Gastos','Receita'])
graf_balanco.set_ylabel('Valores (em bilhões de Reais)')
graf_balanco
```

```python
import json

bairros = []


for obj in json_bairros['features']:
    
    lats = []
    lons = []
    nome = ''
    for elem in obj['geometry']['coordinates'][0]:
        lats.append(elem[1])
        lons.append(elem[0])

    nome = obj['properties']['bairro_nome']
    microrregiao = obj['properties']['microrregiao']
    bairro_codigo = obj['properties']['bairro_codigo']
    
    bairros.append({'nome':nome, 'lat':lats, 'lon':lons, 'microrregiao':microrregiao,
                    'bairro_codigo':bairro_codigo})
    
with open('bairros_mod.json', 'w') as outfile:
    json.dump(bairros, outfile)
```

```python
import geopandas as gpd

json_escolas = gpd.read_file('./dados_prefeitura/educacaomunicipal.geojson')
# print df.columns
json_escolas.head()
```

```python
bairros_rec_geo = gpd.read_file('./dados_prefeitura/bairros.geojson')
bairros_rec_geo.sort_values('bairro_nome').head()
```

```python

pontos_escolas = {'lat':[], 'lon':[], 'id_escola':[], 'qtd_atendidos':[]}

for index, obj in json_escolas.iterrows():
    geo = obj.loc['geometry']
    xs, ys = geo.exterior.xy[0], geo.exterior.xy[1]  
    ident = obj.loc['escola_codigo']
    atendidos = obj.loc['qtd_atendidos']
    
    for ix, pp in enumerate(xs):
        pontos_escolas['lat'].append(xs[ix])
        pontos_escolas['lon'].append(ys[ix])
        pontos_escolas['id_escola'].append(ident)
        pontos_escolas['qtd_atendidos'].append(atendidos)
        break

        
#         print(ix, "(x,y): ", xs[ix], ys[ix])

df_escolas = pd.DataFrame(pontos_escolas)
df_escolas.head()
```

```python
from shapely.geometry import Point
from shapely.geometry.polygon import Polygon

bairro_escola = {'nome_bairro':[], 'id_bairro':[]}

for index, row in df_escolas.iterrows():
    lat = row['lat']
    lon = row['lon']    
    
    for index2, row2 in bairros_rec_geo.iterrows():
        point = Point(lat, lon)
       
        if row2['geometry'].contains(point):
            bairro_escola['nome_bairro'].append(row2['bairro_nome'])
            bairro_escola['id_bairro'].append(row2['bairro_codigo'])
 
```

```python
column_bairros_escola = pd.DataFrame(bairro_escola)
# column_bairros_escola.head()
df_escolas = pd.concat([df_escolas, column_bairros_escola], axis=1)
# print(len(df_escolas))
df_escolas.head()
```

```python
num_escolas_bairro = df_escolas.groupby('nome_bairro').agg({'id_bairro':'count'})

df_agg_escolas_bairro = pd.DataFrame(num_escolas_bairro)
df_agg_escolas_bairro.columns = ['num_escolas']
df_agg_escolas_bairro.index.names = ['bairro_nome']
df_agg_escolas_bairro.reset_index()

# df_agg_escolas_bairro = pd.concat([df_agg_escolas_bairro ])

df_agg_escolas_bairro.sort_values('num_escolas', ascending=False).head()                                                   
```
